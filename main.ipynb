```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('iris.csv')
```

```python
# в качестве признаков выступают длина и ширина лепестка ириса
x = dataset[['petal_length', 'petal_width']]
# в качестве целевой переменной выступает вид цветка
y = dataset['species']
```

```python
# заменяем виды ирисов на числовые представления
y = y.replace({'setosa': 1, 'versicolor': 2, 'virginica': 3})
```

```python
# создаём график, изображающий диаграмму рассеивания
plt.scatter(x['petal_length'], x['petal_width'], c=y, cmap='viridis')

plt.xlabel('petal_length')
plt.ylabel('petal_width')

# создаём цветовую палитру
cbar = plt.colorbar()
cbar.set_label('species')

plt.show()

# создаём модель наивного баейсовского классификатора
model = GaussianNB()

# обучаем модель
model.fit(x, y)

# получаем предсказания модели 
predicted = model.predict([[7, 3], [4.5, 3]])
predicted
```

Из диаграммы рассеивания видна зависимость типа ириса от длины и ширины лепестка. Используя наивный байесовский классификатор мы получили модель, которая определила, что, если длина лепестка цветка меньше 2 см, а ширина находится в пределах одного сантиметра, тогда этот ирис можно отнести к типу setosa. Если длина располагается в промежутке от 3 до примерно 5 сантиметров, а ширина в пределах 1-1.5 сантиметра, тогда это versicolor. Самой большой длиной и шириной лепестка характеризуется тип ириса под названием virginica. Несмотря на свою простоту, в данном случае модель практически идеально классифицировала типы цветов в зависимости от характеристик их лепестков

```python
# сейчас в качестве признаков будут выступать длина и ширина чашелистика
x = dataset[['sepal_length', 'sepal_width']]
# целевая переменная будет та же
y = dataset['species']
```

```python
# снова заменяем типы цветков на числовое представление
y = y.replace({'setosa': 1, 'versicolor': 2, 'virginica': 3})
```

```python
plt.scatter(x['sepal_length'], x['sepal_width'], c=y, cmap='viridis')

plt.xlabel('sepal_length')
plt.ylabel('sepal_width')

cbar = plt.colorbar()
cbar.set_label('species')

plt.show()


model = GaussianNB()

model.fit(x, y)

predicted = model.predict([[7, 3], [4.5, 3]])
predicted
```

В данном случае видно, что обученная модель справилась чуть хуже, чем в прошлый раз. Тип ирисов setosa обладает наименьшей длиной чашелистика, а ширина находится в пределах 3-4.5 см (хотя бывают и исключения). Этот вид цветков легко распознаётся моделью, что нельзя сказать о других двух. На диаграмме видно, что довольно часто жёлтые и зелёные точки находятся в одной области, то есть нельзя провести однозначную и чёткую границу при классификации этих видов ирисов в зависимости от длины и ширины чашелистика. Хотя видно, что versicolor характеризуется меньшими размерами, а в большинстве случаев длина и ширина чашелистика у virginica больше, нельзя сделать однозначный вывод и идеально классифицировать тип ириса по этим параметрам
